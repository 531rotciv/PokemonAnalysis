```python
import pandas as pd
import json
from sqlalchemy import create_engine


```

```python
# Load the JSON data from a file
with open('LA_regional_tournament_data.json') as file:
    data = json.load(file)
```

```python
pokemon_section = data['pokemon']
```

```python
labmaus_la_df = pd.DataFrame(pokemon_section)
```

```python
labmaus_la_df.head(10)
```

```python
two_mon_combination_section = pd.DataFrame(data['composition']['combinations2'])
```

```python
two_mon_combination_section.head(10)
```

```python
pokemon1_name_df = two_mon_combination_section.merge(labmaus_la_df[["id", "pokemon"]], left_on="pokemon1Id", right_on="id").rename(columns={'pokemon':'pokemon1_name'})
```

```python
pokemon1_name_df.head(10)
```

```python
pokemon2_name_df = pokemon1_name_df.merge(labmaus_la_df[["id", "pokemon"]], left_on="pokemon2Id", right_on="id").rename(columns={'pokemon':'pokemon2_name'})
```

```python
pokemon2_name_df.head(10)
```

```python
final_2_mon_composition_df = pokemon2_name_df[["pokemon1_name", "pokemon2_name","usage"]]
```

```python
final_2_mon_composition_df.rename(columns={'usage':'usage_%'}, inplace=True)
```

```python
# Remove the '%' symbol and convert to float
final_2_mon_composition_df["usage_%"] = final_2_mon_composition_df["usage_%"].str.replace('%', '').astype(float)
```

```python
final_2_mon_composition_df.head(20)
```

```python
final_2_mon_composition_df.sort_values(by = "usage_%", ascending = False).head(20)
```

```python
three_mon_combination_section = pd.DataFrame(data['composition']['combinations3'])
pokemon1_name3_df = three_mon_combination_section.merge(labmaus_la_df[["id", "pokemon"]], left_on="pokemon1Id", right_on="id").rename(columns={'pokemon':'pokemon1_name'}).drop(['id_x', 'id_y'], axis=1)
pokemon2_name3_df = pokemon1_name3_df.merge(labmaus_la_df[["id", "pokemon"]], left_on="pokemon2Id", right_on="id").rename(columns={'pokemon':'pokemon2_name'}).drop(['id'], axis=1)
pokemon3_name3_df = pokemon2_name3_df.merge(labmaus_la_df[["id", "pokemon"]], left_on="pokemon3Id", right_on="id").rename(columns={'pokemon':'pokemon3_name'}).drop(['id'], axis=1)
```

```python
pokemon3_name3_df.head(10)
```

```python
pokemon3_name3_df = pokemon3_name3_df[["pokemon1_name","pokemon2_name", "pokemon3_name","usage"]].rename(columns={"usage":"usage_%"})
```

```python
# Remove the '%' symbol and convert to float
pokemon3_name3_df["usage_%"] = pokemon3_name3_df["usage_%"].str.replace('%', '').astype(float)
```

```python
pokemon3_name3_df.sort_values(by = "usage_%", ascending = False).head(20)
```

```python
four_mon_combination_section = pd.DataFrame(data['composition']['combinations4'])
pokemon1_name4_df = four_mon_combination_section.merge(labmaus_la_df[["id", "pokemon"]], left_on="pokemon1Id", right_on="id").rename(columns={'pokemon':'pokemon1_name'}).drop(['id_x', 'id_y'], axis=1)
pokemon2_name4_df = pokemon1_name4_df.merge(labmaus_la_df[["id", "pokemon"]], left_on="pokemon2Id", right_on="id").rename(columns={'pokemon':'pokemon2_name'}).drop(['id'], axis=1)
pokemon3_name4_df = pokemon2_name4_df.merge(labmaus_la_df[["id", "pokemon"]], left_on="pokemon3Id", right_on="id").rename(columns={'pokemon':'pokemon3_name'}).drop(['id'], axis=1)
pokemon4_name4_df = pokemon3_name4_df.merge(labmaus_la_df[["id", "pokemon"]], left_on="pokemon4Id", right_on="id").rename(columns={'pokemon':'pokemon4_name'}).drop(['id'], axis=1)
```

```python
final_4_combinations_df = pokemon4_name4_df[["pokemon1_name","pokemon2_name", "pokemon3_name","pokemon4_name", "usage"]].rename(columns={"usage":"usage_%"})
final_4_combinations_df["usage_%"] = final_4_combinations_df["usage_%"].str.replace('%', '').astype(float)
```

```python
final_4_combinations_df.sort_values(by = "usage_%", ascending = False).head(20)
```

```python
from sqlalchemy import create_engine
engine = create_engine('postgresql://postgres:postgres@localhost:5433/postgres')
final_2_mon_composition_df.to_sql('la_regional_combo_two', engine, schema='pokemon')
```

```python
pokemon3_name3_df.to_sql('la_regional_combo_three', engine, schema='pokemon')
```

```python
final_4_combinations_df.to_sql('la_regional_combo_four', engine, schema='pokemon')
```

```python
final_la_df = labmaus_la_df[["id", "pokemon", "appearances", "usage", "win_rate", "placement", "rank"]].rename(columns={"usage":"usage_%", "win_rate":"win_rate_%"})
final_la_df["usage_%"] = final_la_df["usage_%"].str.replace('%', '').astype(float)
final_la_df["win_rate_%"] = final_la_df["win_rate_%"].str.replace('%', '').astype(float)
final_la_df.to_sql('la_regional_pokemon_usage', engine, schema='pokemon')
```
