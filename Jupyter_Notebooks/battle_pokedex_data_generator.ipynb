```python
def limit_dict(d, max_items=8):
    return dict(list(d.items())[:max_items])
```

```python
pikalytics_df['Abilities'] = pikalytics_df['Abilities'].apply(lambda x: limit_dict(x))
pikalytics_df['Items'] = pikalytics_df['Items'].apply(lambda x: limit_dict(x))
pikalytics_df['Spreads'] = pikalytics_df['Spreads'].apply(lambda x: limit_dict(x))
pikalytics_df['Moves'] = pikalytics_df['Moves'].apply(lambda x: limit_dict(x))
pikalytics_df['Teammates'] = pikalytics_df['Teammates'].apply(lambda x: limit_dict(x))
```

```python
def split_dicts(dictionary,category):
    abilities_flat = {}
    for i, (ability, percentage) in enumerate(dictionary.items(), start=1):
        abilities_flat[f'{category}{i}'] = ability
        abilities_flat[f'Percentage_{category}{i}'] = percentage
    return abilities_flat

```

```python
# Apply the function to the Abilities column
abilities_expanded = pikalytics_df['Abilities'].apply(lambda x: split_dicts(x,'Ability'))
items_expanded = pikalytics_df['Items'].apply(lambda x: split_dicts(x,'Item'))
spreads_expanded = pikalytics_df['Spreads'].apply(lambda x: split_dicts(x,'Spread'))
moves_expanded = pikalytics_df['Moves'].apply(lambda x: split_dicts(x,'Move'))
teammates_expanded = pikalytics_df['Teammates'].apply(lambda x: split_dicts(x,'Teammate'))
```

```python
# Create a DataFrame from the expanded abilities
abilities_df = pd.DataFrame(abilities_expanded.tolist())
items_df = pd.DataFrame(items_expanded.tolist())
spreads_df = pd.DataFrame(spreads_expanded.tolist())
moves_df = pd.DataFrame(moves_expanded.tolist())
teammates_df = pd.DataFrame(teammates_expanded.tolist())
```

```python
items_df.head(5)
```

```python
# Merge the expanded abilities back into the original DataFrame
pikalytics_df = pikalytics_df.drop(columns=['Abilities', 'Items', 'Spreads', 'Moves', 'Teammates']).join(abilities_df)\
.join(items_df)\
.join(spreads_df)\
.join(moves_df)\
.join(teammates_df)
```
