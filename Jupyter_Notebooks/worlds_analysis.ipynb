```python
three_mon_combination_section = pd.DataFrame(data['composition']['combinations3'])
pokemon1_name3_df = three_mon_combination_section.merge(pokemon_section[["id", "pokemon"]], left_on="pokemon1Id", right_on="id").rename(columns={'pokemon':'pokemon1_name'}).drop(['id_x', 'id_y'], axis=1)
pokemon2_name3_df = pokemon1_name3_df.merge(pokemon_section[["id", "pokemon"]], left_on="pokemon2Id", right_on="id").rename(columns={'pokemon':'pokemon2_name'}).drop(['id'], axis=1)
pokemon3_name3_df = pokemon2_name3_df.merge(pokemon_section[["id", "pokemon"]], left_on="pokemon3Id", right_on="id").rename(columns={'pokemon':'pokemon3_name'}).drop(['id'], axis=1)
```

```python
pokemon3_name3_df = pokemon3_name3_df[["pokemon1_name","pokemon2_name", "pokemon3_name","usage"]].rename(columns={"usage":"usage_%"})
pokemon3_name3_df["usage_%"] = pokemon3_name3_df["usage_%"].str.replace('%', '').astype(float)
```

```python
pokemon3_name3_df.sort_values(by = "usage_%", ascending = False).head(20)
```

```python
four_mon_combination_section = pd.DataFrame(data['composition']['combinations4'])
pokemon1_name4_df = four_mon_combination_section.merge(pokemon_section[["id", "pokemon"]], left_on="pokemon1Id", right_on="id").rename(columns={'pokemon':'pokemon1_name'}).drop(['id_x', 'id_y'], axis=1)
pokemon2_name4_df = pokemon1_name4_df.merge(pokemon_section[["id", "pokemon"]], left_on="pokemon2Id", right_on="id").rename(columns={'pokemon':'pokemon2_name'}).drop(['id'], axis=1)
pokemon3_name4_df = pokemon2_name4_df.merge(pokemon_section[["id", "pokemon"]], left_on="pokemon3Id", right_on="id").rename(columns={'pokemon':'pokemon3_name'}).drop(['id'], axis=1)
pokemon4_name4_df = pokemon3_name4_df.merge(pokemon_section[["id", "pokemon"]], left_on="pokemon4Id", right_on="id").rename(columns={'pokemon':'pokemon4_name'}).drop(['id'], axis=1)
```

```python
import re
def remove_brackets(text):
    return re.sub(r'\[(.*)\]', r'\1', str(text))
```

```python

# Remove brackets from each Pokémon column
columns_to_clean = ["pokemon1", "pokemon2", "pokemon3", "pokemon4", "pokemon5", "pokemon6"]

for col in columns_to_clean:
    worlds_teams_df[col] = worlds_teams_df[col].apply(remove_brackets)
    worlds_teams_df[col] = worlds_teams_df[col].str.replace("'", "", regex=False)


# Now the columns will no longer have brackets around the names.
```
